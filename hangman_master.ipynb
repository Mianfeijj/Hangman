```python
import random
from hangman_word import word_list 
from hangman_graphic import stages, logo
import os

def clear():
    os.system('clear')

print (logo)

chosen_word = random.choice (word_list)
word_length = len(chosen_word)
lives = len (stages) - 1

answer_blanks = []
for i in range (word_length):
    answer_blanks += '_'
print(f"{' '.join(answer_blanks)}")

game_over = False
while game_over is False:

    guess = input('Guess a letter: ').lower()
    
    clear()
    
    if guess in answer_blanks:
        print (f'Letter {guess} is already there. Try another letter.')

    for p in range (word_length):
        if chosen_word[p] == guess:
            answer_blanks[p] = guess
    
    print(f"{' '.join(answer_blanks)}")
    
    if guess not in chosen_word:
        lives -= 1
        print ('Wrong answer, you lost a life.')
        
    if '_' not in answer_blanks:
        game_over = True
        print ('You win!')
    elif lives == 0:
        game_over = True
        print ('You loose!')
        print (f'The answer is {chosen_word}')
    
    print (stages[lives])
```
